  <h2>🪪 Smart Bathroom Passes – Clean, Efficient, and Modern</h2>
  <p>One of the most innovative ideas we saw was at the CSA showcase. Instead of using a traditional bathroom pass, students can now just show the barcode on their ID cards. It's a simple but genius solution—reducing the spread of germs and keeping everything neat and tidy.</p>

  <h2>🎓 HPOE Showcase – Learning Beyond the Books</h2>
  <p>This HPOE (Health Professions and Occupations Education) Showcase from another school impressed us with its integration of VEX robotics and other hands-on tools. It’s amazing to see education become more interactive and inspiring for future scientists, engineers, and health professionals.</p>

  <h2>🔧 Poway High’s Automotive Engineering Program</h2>
  <p>At Poway High School, students aren't just reading about cars—they’re building them. Through the school’s automotive class, learners gain hands-on experience constructing a car from the ground up, diving deep into topics like bumpers, wiring, and structural integrity. A real-world learning environment that drives curiosity!</p>
