    <h3>🔢 What is Binary?</h3>
    <p>Binary is a Base-2 number system made up of only 0 and 1. Computers process binary to represent all types of data: text, images, video, music, and logic.</p>

      <li>One byte = 8 bits → Can represent 256 values</li>

      <li>Music and videos are stored and transmitted as binary streams</li>

    <h2>🔁 SECTION 2: Converting Between Binary & Decimal</h2>
    <h3>🔄 Decimal to Binary (Divide by 2 Method)</h3>
    <p><strong>Steps:</strong></p>
    <ol>
      <li>Divide the number by 2</li>
      <li>Record the remainder</li>
      <li>Repeat until the quotient is 0</li>
      <li>Read the remainders bottom to top</li>
    </ol>
    <h3>Example: Convert 25 to Binary</h3>
    <p>25 ÷ 2 = 12 R1<br>
    12 ÷ 2 = 6 R0<br>
    6 ÷ 2 = 3 R0<br>
    3 ÷ 2 = 1 R1<br>
    1 ÷ 2 = 0 R1<br>
    ➡ Binary: 11001</p>

    <h3>🔁 Binary to Decimal</h3>

    <h2>🔎 SECTION 3: Binary Search vs Linear Search</h2>
    <h3>📘 What is Binary Search?</h3>
    <p>Binary Search is a powerful algorithm used to quickly find an element in a sorted list by repeatedly dividing the search interval in half.</p>
    <p>Binary search is much faster than linear search for large datasets. It’s often referred to as a “divide and conquer” algorithm.</p>
    <h4>Steps:</h4>
    <ol>
      <li>Check the middle element</li>
      <li>If it’s the target, return it</li>
      <li>If not, search the left or right half</li>
      <li>Repeat until found or empty</li>
    </ol>
    <h3>⚠️ Limitations of Binary Search</h3>
    <ul>
      <li>🚫 List must be sorted</li>
      <li>😵 Not useful for very small datasets</li>
      <li>🧩 A bit more complex to code</li>
    </ul>

    <h2>⏱️ Time Complexity of Binary Search</h2>
    <p>Binary search is efficient because it divides the search space in half with every step. Here’s how its time complexity breaks down:</p>
    <ul>
      <li>🧠 At each step, we eliminate half of the remaining elements.</li>
      <li>🧮 This “halving” continues until there’s only one element left.</li>
      <li>✨ This gives us a logarithmic number of steps — specifically:</li>
      <li>Time Complexity: O(log₂ n) where n is the number of elements in the list.</li>
    </ul>
    <p>✅ So in the worst case, binary search takes about log₂(n) steps to find the target (or determine it’s not there).</p>
    <p>This makes it much faster than linear search (O(n)), especially on large datasets!</p>

    <h3>🐍 Binary Search in Python</h3>
    <div class="code-block">
      <pre>
# Define the binary search function
def binary_search(arr, target):
    # Set the left and right bounds of the list
    left, right = 0, len(arr) - 1

    # Continue the loop while the bounds have not crossed
    while left <= right:
        # Find the middle index
        mid = (left + right) // 2
        print(f"Checking index {mid}: {arr[mid]}")  # Debug print

        # If the middle value is the target, return the index
        if arr[mid] == target:
            return mid
        # If the target is greater, search the right half
        elif arr[mid] < target:
            left = mid + 1
        # If the target is smaller, search the left half
        else:
            right = mid - 1

    return -1  # If not found, return -1

# Example list
numbers = [1, 3, 5, 7, 9, 11, 13]

# Search for the number 7 in the list
print(binary_search(numbers, 7))  # Output: 3
      </pre>

// Define the binary search function
function binarySearch(arr, target) {
    // Set the left and right bounds of the list
    let left = 0;
    let right = arr.length - 1;

    // Loop while the bounds haven't crossed
    while (left <= right) {
        // Find the middle index
        let mid = Math.floor((left + right) / 2);
        console.log("Checking index:", mid, arr[mid]); // Debug print

        // If the middle value is the target, return the index
        if (arr[mid] === target) {
            return mid;
        // If the target is greater, search the right half
        } else if (arr[mid] < target) {
            left = mid + 1;
        // If the target is smaller, search the left half
        } else {
            right = mid - 1;
        }
    }

    // If the target is not found, return -1
    return -1;
}

// Example list
let numbers = [1, 3, 5, 7, 9, 11, 13];

// Search for the number 7 in the list
console.log(binarySearch(numbers, 7)); // Output: 3
