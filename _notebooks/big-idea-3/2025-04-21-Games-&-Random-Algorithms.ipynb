    <h2>üèÄ Bouncing Ball Simulation (No Randomness)</h2>
    <p>Uses physics formulas to simulate a ball bouncing under gravity. Run this part in a Jupyter Notebook to see animation:</p>
    <pre><code>import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML
# Simulation parameters
gravity = -9.8
step = 0.05
restitution = 0.75
y = 10
v = 0
def update_physics(y, v):
    v += gravity * step
    y += v * step
    if y <= 0:
        y = 0
        v = -v * restitution
    return y, v

fig, ax = plt.subplots()
ball, = plt.plot([], [], 'o', markersize=20)
ax.set_xlim(0, 1)
ax.set_ylim(0, 12)
ax.set_title("Bouncing Ball Simulation")
ax.get_xaxis().set_visible(False)
ax.get_yaxis().set_visible(False)

def init():
    ball.set_data([0.5], [y])
    return ball,

def animate(i):
    global y, v
    y, v = update_physics(y, v)
    ball.set_data([0.5], [y])
    return ball,

ani = animation.FuncAnimation(fig, animate, init_func=init,
                              frames=200, interval=50, blit=True)

HTML(ani.to_jshtml())</code></pre>
