    <section class="bg-white shadow-md rounded-xl p-6">
      <h2 class="text-2xl font-semibold text-indigo-600 mb-4">Legal Concerns</h2>
      <h3 class="text-xl font-medium text-gray-700 mb-2">Intellectual Property (IP)</h3>
      <ul class="list-disc pl-5 mb-4 space-y-1">
        <li>IP includes inventions, art, code, logos, etc.</li>
        <li>Digital content is easy to copy and distribute.</li>
        <li>Copyright protects original works like code, music, books.</li>
        <li>Other protections: <strong>Patents</strong>, <strong>Trademarks</strong>, <strong>Trade Secrets</strong></li>
      </ul>
      <h3 class="text-xl font-medium text-gray-700 mb-2">Enforcing IP</h3>
      <ul class="list-disc pl-5 space-y-1">
        <li>Use licenses and DRM to control how code is used.</li>
        <li>Monitor and legally protect your rights.</li>
      </ul>
    </section>
    <section class="bg-white shadow-md rounded-xl p-6">
      <h2 class="text-2xl font-semibold text-indigo-600 mb-4">Licensing Code</h2>
      <p class="mb-4">By default, all rights are reserved unless a license is added. Open-source licenses define what others can legally do with your code.</p>
      <h3 class="text-xl font-medium text-gray-700 mb-2">Common Open-Source Licenses</h3>
      <div class="overflow-x-auto">
        <table class="w-full text-left border-collapse border border-gray-300 text-sm">
          <thead class="bg-gray-100">
            <tr>
              <th class="border p-2">License</th>
              <th class="border p-2">Permissions</th>
              <th class="border p-2">Restrictions</th>
              <th class="border p-2">Common Uses</th>
            </tr>
          </thead>
          <tbody>
            <tr>
              <td class="border p-2">MIT</td>
              <td class="border p-2">Use, modify, share with credit</td>
              <td class="border p-2">No liability/warranty</td>
              <td class="border p-2">Web apps, libraries</td>
            </tr>
            <tr>
              <td class="border p-2">Apache 2.0</td>
              <td class="border p-2">Like MIT + patent protection</td>
              <td class="border p-2">Must include license notice</td>
              <td class="border p-2">APIs, enterprise tools</td>
            </tr>
            <tr>
              <td class="border p-2">GPL</td>
              <td class="border p-2">Stay open-source</td>
              <td class="border p-2">Can’t privatize mods</td>
              <td class="border p-2">Non-profits, open tools</td>
            </tr>
            <tr>
              <td class="border p-2">BSD</td>
              <td class="border p-2">Use, modify, redistribute</td>
              <td class="border p-2">No endorsement</td>
              <td class="border p-2">Academic projects</td>
            </tr>
            <tr>
              <td class="border p-2">Creative Commons</td>
              <td class="border p-2">Flexible for non-code</td>
              <td class="border p-2">Not for software</td>
              <td class="border p-2">Docs, art, media</td>
            </tr>
          </tbody>
        </table>
      </div>
    </section>
    <section class="bg-white shadow-md rounded-xl p-6">
      <h2 class="text-2xl font-semibold text-indigo-600 mb-2">Breaking the Law</h2>
      <ul class="list-disc pl-5 space-y-1">
        <li>Using code without a license = copyright violation</li>
        <li>Modifying GPL code without sharing changes = illegal</li>
        <li>Removing license or credit = violation</li>
        <li>Selling open-source code without permission = illegal</li>
      </ul>
    </section>
    <section class="bg-white shadow-md rounded-xl p-6">
      <h2 class="text-2xl font-semibold text-indigo-600 mb-2">Ethical Concerns</h2>
      <ul class="list-disc pl-5 mb-4 space-y-1">
        <li>Using code without attribution = plagiarism</li>
        <li>Open-source relies on trust, credit, and fairness</li>
        <li>Profiting from others' work without credit is unethical</li>
      </ul>
      <h3 class="text-xl font-medium text-gray-700 mb-2">Example: Google vs Oracle</h3>
      <p class="mb-4">Oracle accused Google of stealing Java APIs in Android. Google argued it was fair use. The case went all the way to the Supreme Court.</p>
      <h3 class="text-xl font-medium text-gray-700 mb-2">Plagiarism vs. Proper Use</h3>
      <ul class="list-disc pl-5 space-y-1">
        <li><strong>Wrong:</strong> Copy-pasting from Stack Overflow without credit</li>
        <li><strong>Right:</strong> Forking repos, keeping licenses, giving credit</li>
      </ul>
    </section>
    <section class="bg-white shadow-md rounded-xl p-6">
      <h2 class="text-2xl font-semibold text-indigo-600 mb-2">Homework</h2>
      <ol class="list-decimal pl-5 space-y-1">
        <li>Go to your GitHub repo → Click “Add file” → “Create new file” → Name it <code>LICENSE</code></li>
        <li>Select a license template and commit</li>
        <li>Explain in your README why you chose that license</li>
      </ol>
